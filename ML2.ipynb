```python
import pandas as pd

# Create sample dataset
data = {
    "Name": ["Aman", "Riya", None, "Sneha", "Karan"],
    "Age": [20, 21, 19, None, 22],
    "Marks": [85, None, 78, 92, 88],
    "Department": ["CS", "IT", "CS", "IT", None]
}

df = pd.DataFrame(data)
df.to_csv("data.csv", index=False)

df
```

```python
df = pd.read_csv("data.csv")
```

```python
print("Experiment 1")
print(df.isnull())
print("\nMissing values count:\n", df.isnull().sum())
```

```python
print("\nExperiment 2")
print("Before cleaning:", df.shape)

df_clean = df.dropna()

print("After cleaning:", df_clean.shape)
```

```python
print("\nExperiment 3")

df_fill = df.copy()

# Fill numerical
df_fill["Age"].fillna(df["Age"].mean(), inplace=True)
df_fill["Marks"].fillna(df["Marks"].median(), inplace=True)

# Fill categorical
df_fill["Department"].fillna("Unknown", inplace=True)

print(df_fill)
```

```python
print("\nExperiment 4")

print("Before:\n", df_fill.dtypes)

df_fill["Age"] = df_fill["Age"].astype(int)

print("\nAfter:\n", df_fill.dtypes)
```

```python
print("\nExperiment 5")

df_rename = df_fill.rename(columns={"Marks": "Score"})

print(df_rename)
```

```python
print("\nExperiment 6")

df_replace = df_rename.copy()
df_replace["Department"].replace("Unknown", "CS", inplace=True)

print(df_replace)
```

```python
import matplotlib.pyplot as plt

print("\nExperiment 7")

plt.figure()
df_replace["Score"].plot(kind='line', title="Line Plot of Score")
plt.show()

plt.figure()
df_replace["Department"].value_counts().plot(kind='bar', title="Department Count")
plt.show()
```

```python
print("\nExperiment 8")

plt.figure()
plt.scatter(df_replace["Age"], df_replace["Score"])
plt.xlabel("Age")
plt.ylabel("Score")
plt.title("Age vs Score")
plt.show()
```

```python
# ------------------------------------------
print("\nExperiment 9")

plt.figure()
plt.hist(df_replace["Score"])
plt.title("Score Distribution")
plt.show()
```

```python
print("\nExperiment 10")

plt.figure()
plt.boxplot(df_replace["Score"])
plt.title("Boxplot of Score")
plt.show()
```
